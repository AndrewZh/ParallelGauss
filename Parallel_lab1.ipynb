```python
data = pd.read_csv("LinearGauss/reports/2015-05-22/NxN_4_nodes.csv")
data
```

```python
fig = figure(figsize=(15, 15))
plot(data["Size"].values, data["Linear Gauss"].values, marker = 'o', c = 'magenta', alpha = 0.6, label = "Linear")
plot(data["Size"].values, data["Tiled Single"].values, marker = '<', c = 'blue', alpha = 0.6, label = "Tiled")
plot(data["Size"].values, data["OpenMP(outer)"].values, marker = '>', c = 'yellow', alpha = 0.6, label = "Outer")
plot(data["Size"].values, data["OpenMP(inner)"].values, marker = 'p', c = 'red', alpha = 0.6, label = "Inner")
plot(data["Size"].values, data["OpenMP(single loop)"].values, marker = 'p', c = 'red', alpha = 0.6, label = "Single loop")
legend(loc = 2)
```
